# **I. LIBRARIES**

````python
import pandas as pd
import numpy as np
import pickle
import random
import matplotlib.pyplot as plt
````

# **II. PIPELINE LOADING**

````python
with open('pipeline_yosef.pkl', 'rb') as file:
  pipeline_yosef = pickle.load(file)
````

````python
pipeline_yosef
````

# **III. DATA INFERENCE**

````python
# seed_value = 666
# random.seed(seed_value)
# np.random.seed(seed_value)

al = np.random.uniform(0,5.05,11111)
ars = np.random.uniform(0,1.05,11111)
ba = np.random.uniform(0,4.94,11111)
cd = np.random.uniform(0,0.13,11111)
r2ncl = np.random.uniform(0,8.68,11111)
cr = np.random.uniform(0,0.90,11111)
v = np.random.uniform(0,1.00,11111)
no2 = np.random.uniform(0,2.93,11111)
clo4 = np.random.uniform(0,60.01,11111)
ra = np.random.uniform(0,7.99,11111)
ag = np.random.uniform(0,0.50,11111)
u = np.random.uniform(0,0.09,11111)



data_inf = pd.DataFrame({
    'alumunium' : al,
    'arsenic': ars,
    'barium': ba,
    'cadmium' : cd,
    'chloramine': r2ncl,
    'chromium': cr,
    'viruses': v,
    'nitrites': no2,
    'perchlorate': clo4,
    'radium': ra,
    'silver': ag,
    'uranium': u
})

data_inf
````

# **IV. PREDICT**

````python
y_pred_inf = pipeline_yosef.predict(data_inf)
y_pred_inf_df = pd.DataFrame(y_pred_inf,columns=['is_safe_pred'])

inf_table = pd.concat([data_inf,y_pred_inf_df],axis=1)
inf_table
````

````python
# Plotbar for 0 and 1 label in `is_safe`

def_count = (inf_table.is_safe_pred.value_counts(normalize=True)*100)
def_count.plot.bar(figsize=(3,4))
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12)
plt.title("Water is Safe", fontsize=15)
for x,y in zip([0,1],def_count):
    plt.text(x, y, '{:.2f}%'.format(y), fontsize=12)
plt.show()
````

There is verry lil' data classified as 1 (safe), the effect from extremely imbalance original data I use.
